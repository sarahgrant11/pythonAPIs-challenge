```python
API_key = "4343bc3726df924a38d9c0897b39a1a9"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import requests
from citipy import citipy

lat_range = (-90,90)
lon_range = (-180,180)
```

```python
lat_lon = []
cities = []

# create random combinations of latitude and longitude 

random_lat = np.random.uniform(lat_range[0],lat_range[1], size = 1500)
random_lon = np.random.uniform(lon_range[0], lon_range[1], size = 1500)

lat_lon = zip(random_lat, random_lon)

for i in lat_lon:
    city = citipy.nearest_city(i[0],i[1]).city_name

    if city not in cities:
        cities.append(city)

print(len(cities))

```

```python

URL = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + API_key 
city_data = []


print("starting API call")
print("-------------")

city_count = 1
set_count = 1

#Starting API loop
for i , city in enumerate(cities):

    if (i%50 == 0 and i >= 50):
        set_count += 1 
        city_count = 0 
    
    #create unique URL for API call 
    city_URL = URL + "&q=" + city

    print("Processing Record Number %s of Set %s | %s" % (city_count, set_count, city))

    city_count += 1 

    #grabbing relevant information from API call 
    try: 
        city_weather = requests.get(city_URL).json()

        # Parse out the max temp, humidity, and cloudiness
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_date = city_weather["dt"]

        #append city data list 
        city_data.append({"City Name": city, 
                          "Lat": city_lat, 
                          "Lng": city_lng, 
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

    except:
        print("City Not Found...")
        pass
```
